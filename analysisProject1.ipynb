# Which age group shows the most interest in Computer IT courses?

# select Age column from the data set

```python
data = pd.read_csv("cleaned-major2024.csv") # read data file 

import pandas as pd
import matplotlib.pyplot as plt

data # cleaned data frame from non-major2024 
age_column = data['Age '] # select column Age in the data frame
print(age_column) #print age column 
```

```python
# Create a DataFrame with the given age group data
data1 = {
    'Age Group': ['18 and younger', '19-20', '21-24', '25-34', '35-64'],
    'Interest': [14, 44, 17, 11, 6]
}

df = pd.DataFrame(data1)

# Sort values by interest for better visualization
df = df.sort_values(by='Interest', ascending=False)

# Plot the data using a bar chart
plt.figure(figsize=(8, 5))
plt.bar(df['Age Group'], df['Interest'], color='blue', alpha=0.7)
plt.title('Interest in Computer IT Courses by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Interested Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Explanation
most_interested_age_group = df.loc[df['Interest'].idxmax(), 'Age Group']
most_interest_count = df['Interest'].max()

print(f"The {most_interested_age_group} age group shows the most interest in Computer IT courses, with {most_interest_count} students.")

```

The 19-20 age group shows the most interest in Computer IT courses, with 44 students. To attract more young people, CCM should develop additional programs and events targeting high school and middle school students.

# What are the top 5 reasons that have the greatest impact on students' decision to attend the CCM?

```python
data = pd.read_csv("cleaned-major2024.csv") # read data file 

import pandas as pd
import matplotlib.pyplot as plt


data # cleaned data frame from non-major2024 

# List of columns to analyze
columns_to_count = [
    'impact_transfer_student',
    'impact_online_offerings',
    'impact_college_reputation',
    'impact_choice_programs',
    'impact_affordable_cost',
    'impact_family_friend_influence'
]

# Create a dictionary to hold counts for each column
count_results = {}

# Loop through each column and count unique values
for column in columns_to_count:
    count_results[column] = data[column].value_counts()

# Print the count results
for column, counts in count_results.items():
    print(f"\nCounts for column '{column}':")
    print(counts)
 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary of the impact data for each column
data = {
    'impact_transfer_student': {'No Impact': 50, 'High Impact': 20, 'Some Impact': 13},
    'impact_online_offerings': {'High Impact': 37, 'No Impact': 34, 'Some Impact': 16},
    'impact_college_reputation': {'Some Impact': 36, 'High Impact': 28, 'No Impact': 22},
    'impact_choice_programs': {'High Impact': 66, 'Some Impact': 13, 'No Impact': 10},
    'impact_affordable_cost': {'High Impact': 71, 'Some Impact': 17, 'No Impact': 1},
    'impact_family_friend_influence': {'No Impact': 43, 'Some Impact': 25, 'High Impact': 12}
}

# Convert the data into a DataFrame for better manipulation
df = pd.DataFrame(data)

# Plot the data
df.T.plot(kind='bar', figsize=(10, 7), width=0.7)
plt.title('Impact Factors on Student Decisions')
plt.xlabel('Factors')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Impact Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The top five factors that have the greatest impact on students' decisions to attend CCM are affordable cost, choice of programs, online offerings, college reputation, and transfer opportunities. Focusing on these areas can help CCM enhance its recruitment process and attract more students.

```python
data = pd.read_csv("cleaned-major2024.csv") # read data file 

import pandas as pd
import matplotlib.pyplot as plt


data # cleaned data frame from non-major2024 

# List of columns to analyze
columns_to_count = [
    'interested_computing_App_Dev', 'interested_computing_Comp_Prog', 'interested_computing_CyberSec']
    

# Create a dictionary to hold counts for each column
count_results = {}

# Loop through each column and count unique values
for column in columns_to_count:
    count_results[column] = data[column].value_counts()

# Print the count results
for column, counts in count_results.items():
    print(f"\nCounts for column '{column}':")
    print(counts)
```

```python
import matplotlib.pyplot as plt

# Data for student interest in each course
courses = ['Cyber Security', 'App Development', 'Computer Programming']
yes_counts = [31, 24, 22]
no_counts = [61, 68, 70]

# Create a bar graph for "Yes" responses
plt.figure(figsize=(8, 5))
plt.barh(courses, yes_counts, color='blue', label='Yes')
plt.barh(courses, no_counts, left=yes_counts, color='lightgrey', label='No')

# Add title and labels
plt.title('Student Interest in IT Courses')
plt.xlabel('Number of Students')
plt.ylabel('Course')
plt.legend(title='Response')

# Display the bar graph
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for gender counts
data = {
    'Gender': ['Woman', 'Man', 'Prefer not to say', 'Non-binary'],
    'Count': [60, 29, 2, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the data using a pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    df['Count'],
    labels=df['Gender'],
    autopct='%0.1f%%',  # Display percentages
    startangle=90,  # Start the pie chart at 90 degrees
    colors=['red', 'blue', 'orange', 'purple'],  # Colors for each category
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}  # Define edge color for clarity
)
plt.title('Gender Distribution')
plt.show()
```

## The survey results show that the majority of respondents are women (60), followed by men (29). A smaller number of respondents chose 'Prefer not to say' (2) and 'Non-binary' (1). The significant difference lies in the higher number of women compared to other gender categories. To better support this demographic, CCM should develop more programs that make it easier for women to pursue their studies while managing other responsibilities.
