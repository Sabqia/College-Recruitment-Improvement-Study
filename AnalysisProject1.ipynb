# Which age group shows the most interest in Computer IT courses?

```python


# Count the frequency of each age group
age_counts = data['Age '].value_counts()

# Identify the index of the age group with the most interest
max_age_group = age_counts.idxmax()

# Set bar colors: blue for the max age group, light color for others
colors = ['lightgray' if age != max_age_group else 'blue' for age in age_counts.index]

# Plot the data
plt.figure(figsize=(10, 6))
age_counts.plot(kind='bar', color=colors)
plt.title('Interest in Computer IT Courses by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

# What are the top 5 reasons that have the greatest impact on students' decision to attend the CCM?

```python
 # List of columns to analyze
columns_to_count = [
    'impact_transfer_student',
    'impact_online_offerings',
    'impact_college_reputation',
    'impact_choice_programs',
    'impact_affordable_cost',
    'impact_family_friend_influence'
]

# Create a dictionary to hold counts for each column
count_results = {}

# Loop through each column and count unique values
for column in columns_to_count:
    count_results[column] = data[column].value_counts().reset_index()

# Print the count results
for column, counts in count_results.items():
    print(f"\nCounts for column '{column}':")
    print(counts)
 
```

```python
# plot Analysis
# List of columns to analyze
columns_to_count = [
    'impact_transfer_student',
    'impact_online_offerings',
    'impact_college_reputation',
    'impact_choice_programs',
    'impact_affordable_cost',
    'impact_family_friend_influence'
]

# Dictionary to hold counts of 'High Impact' values
high_impact_counts = {}

# Loop through each column to count 'High Impact' values
for column in columns_to_count:
    counts = data[column].value_counts()
    high_impact_count = counts.get("High Impact", 0)  # Get count of 'High Impact', or 0 if it doesn't exist
    high_impact_counts[column] = high_impact_count

# Convert dictionary to DataFrame for easier plotting
impact_df = pd.DataFrame(list(high_impact_counts.items()), columns=['Reason', 'High Impact Count'])

# Sort by 'High Impact Count' in descending order and select top 5
top_5_impact = impact_df.sort_values(by='High Impact Count', ascending=False).head(5)

# Map column names to more descriptive labels
top_5_impact['Reason'] = top_5_impact['Reason'].map({
    'impact_transfer_student': 'Transfer Opportunities',
    'impact_online_offerings': 'Online Offerings',
    'impact_college_reputation': 'College Reputation',
    'impact_choice_programs': 'Choice of Programs',
    'impact_affordable_cost': 'Affordable Cost',
    'impact_family_friend_influence': 'Family/Friend Influence'
})

# Plot the top 5 reasons
plt.figure(figsize=(10, 6))
plt.barh(top_5_impact['Reason'], top_5_impact['High Impact Count'], color='skyblue')
plt.xlabel('Number of Students (High Impact)')
plt.title('Top 5 Reasons Impacting Studentsâ€™ Decision to Attend CCM')
plt.gca().invert_yaxis()  # Invert y-axis for descending order
plt.show()
```

The top five factors that have the greatest impact on students' decisions to attend CCM are affordable cost, choice of programs, online offerings, college reputation, and transfer opportunities. Focusing on these areas can help CCM enhance its recruitment process and attract more students.

```python
# Data for gender counts
data = {
    'Gender': ['Woman', 'Man', 'Prefer not to say', 'Non-binary'],
    'Count': [60, 29, 2, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the data using a bar chart
plt.figure(figsize=(8, 5))
plt.bar(df['Gender'], df['Count'], color=['blue', 'green', 'orange', 'purple'], alpha=0.7)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Respondents')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Pie Graph
# Data for gender counts
data = {
    'Gender': ['Woman', 'Man', 'Prefer not to say', 'Non-binary'],
    'Count': [60, 29, 2, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Define explode parameter to slightly separate the slices
explode = (0.1, 0.1, 0.1, 0.1)  # Explode all slices

# Plot the data using a pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    df['Count'],
    explode=explode,  # Explode the slices
    autopct='%0.1f%%',  # Display percentages
    startangle=90,  # Start the pie chart at 90 degrees
    colors=['red', 'blue', 'orange', 'purple'],  # Colors for each category
    wedgeprops={'edgecolor': 'black', 'linewidth': 1},  # Define edge color for clarity
    shadow=True  # Add shadow for better visibility
)

# Increase font size for percentages
plt.setp(autotexts, fontsize=12, color='white', weight='bold')

# Set title
plt.title('Gender Distribution', fontsize=16)

# Add legend
plt.legend(wedges, df['Gender'], title='Gender', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

## The survey results show that the majority of respondents are women (60), followed by men (29). A smaller number of respondents chose 'Prefer not to say' (2) and 'Non-binary' (1). The significant difference lies in the higher number of women compared to other gender categories. To better support this demographic, CCM should develop more programs that make it easier for women to pursue their studies while managing other responsibilities.
